```python
import subprocess
import re
import pandas as pd
import matplotlib.pyplot as plt

TIME_REGEX = re.compile(r'^\s+(?P<real>[0-9.]+) real\s+(?P<user>[0-9.]+) user\s+(?P<sys>[0-9.]+) sys$', re.MULTILINE)
SPACE_REGEX = re.compile(r'(?P<space_in_bytes>\d+)\s+maximum resident set size', re.MULTILINE)
```

```python
data = {}

dict_implementations = (
    ('chaining', './soln_3_chaining_hashmap'),
    ('open addressing', './soln_4_open_addressing_hashmap'),
)

for implementation_name, exe in dict_implementations:
    
    data[implementation_name] = []
    
    for dict_capacity in map(int, (1.2e8, 1.4e8, 2e8, 4e8)):
        
#         if (implementation_name == 'chaining' and dict_capacity > 2e8):
            # in this case, space used > 6GB and machine starts thrashing{}
#             next
        
        out = subprocess.check_output(['/usr/bin/time', '-l',
                                       exe, 'usernames_100M.txt', str(dict_capacity)],
                                      stderr=subprocess.STDOUT)
        time_match = TIME_REGEX.search(out)
        space_match = SPACE_REGEX.search(out)
        row = {
            'implementation': implementation_name,
            'dict_capacity': dict_capacity,
            'time_real': float(time_match.group('real')),
            'time_user': float(time_match.group('user')),
            'time_sys': float(time_match.group('sys')),
            'space': int(space_match.group('space_in_bytes')),
        }
        data[implementation_name].append(row)
    
    data[implementation_name] = pd.DataFrame(data[implementation_name])
```

```python
data
```

```python
for implementation, df in data.items():
    plt.plot(df.dict_capacity, df.time_real,
             marker='o', label=implementation)
axes = plt.gca()
# axes.set_xlim(xmin=0, xmax=3.2e8)
axes.set_ylim(ymin=0)#, ymax=10)
plt.xlabel('Hashmap capacity')
plt.ylabel('Real time (seconds)')
plt.legend()
plt.show()

for implementation, df in data.items():
    plt.plot(df.dict_capacity, df.space,
             marker='o', label=implementation)
axes = plt.gca()
# axes.set_xlim(xmin=0, xmax=3.2e8)
axes.set_ylim(ymin=0)#, ymax=6)
plt.xlabel('Hashmap capacity')
plt.ylabel('Maximum resident set size (bytes)')
plt.legend()
plt.show()
```
